## 2. Merging and Selecting Features
The `credits` dataframe contains our cast and crew, but we need to join it with the main `movies` dataframe. After merging, we will drop columns (like budget or homepage) that do not contribute to semantic text similarity.

```python
movies = movies.merge(credits,left_on='id', right_on='movie_id') #Merge on the movie title or ID
```

```python
movies.drop(columns=['movie_id','title_y'], inplace=True)
movies.rename(columns={'title_x':'title'}, inplace=True)
# Since both tables have a 'title' column, Pandas will rename them to 'title_x' and 'title_y'. 
# We can drop the extra ID and title columns to keep it clean:
```

```python
# Keep only the features that contribute to "content" similarity
movies = movies[['id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
# Drop rows with missing critical data (like an empty overview)
movies.dropna(inplace=True)
```
